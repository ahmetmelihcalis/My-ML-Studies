```python
plt.scatter("Height","Weight",data = df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlete Weight vs Height")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x = "Height",y = "Weight", data =df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athletes Height and Weight")
plt.show()
```

```python
sns.set_style("darkgrid")

ax = sns.scatterplot(x="Height", y="Weight", data=df)
ax.set_title("Athletes Height and Weight")
ax.set_xlabel("Height")
ax.set_ylabel("Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height",y = "Weight", hue = "Sex",data =df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athletes Height and Weight")
plt.show()
```

```python
df["Medal"].unique()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height",y = "Weight", hue = "Sex", style = "Medal",size = "Age",data =df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athletes Height and Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x = "Height",y = "Weight", hue = "Sex",data =df)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athletes Height and Weight")
plt.show()
```

```python
sns.set_style("dark")
sns.displot(x="Height", hue = "Sex", data = df)
plt.ylabel("Frequency")
plt.title("Athletes Height Distrubution")
plt.show()
```

```python
sns.set_style("dark")
sns.displot(x="Height", hue = "Sex", data = df,kind = "kde")
plt.ylabel("Frequency")
plt.title("Athletes Height Distrubution")
plt.show()
```

```python
sns.barplot(x="Medal",y = "Height", hue = "Sex",data = df)
plt.title("Medals by Height")
plt.show()
```

```python
sns.catplot(x="Medal",y="Height",hue="Sex",col="Season",data=df)
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data Value")
plt.grid(True)
plt.show()
```

```python
dataframe = sns.load_dataset("titanic")
dataframe
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
sns.boxplot(x="class",y="age",data = dataframe)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x="class",y="fare",data = dataframe)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.tight_layout
plt.show()
```
